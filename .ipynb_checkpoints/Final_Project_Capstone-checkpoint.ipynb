```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
Tor_venues = getNearbyVenues(names=DF['Neighbourhood'],
                                   latitudes=DF['Latitude'],
                                   longitudes=DF['Longitude']
                                  )
```

```python
Tor_venues.head()
```

```python
# one hot encoding
Tor_venues_onehot = pd.get_dummies(Tor_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
Tor_venues_onehot['Neighborhood'] = Tor_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [Tor_venues_onehot.columns[-1]] + list(Tor_venues_onehot.columns[:-1])
Tor_venues_onehot = Tor_venues_onehot[fixed_columns]

Tor_venues_onehot.head()
```

```python
Tor_venues_grouped = Tor_venues_onehot.groupby('Neighborhood').sum()
Tor_venues_grouped = Tor_venues_grouped.astype(float)
```

```python
Tor_venues_grouped['Total'] = Tor_venues_grouped[0:].sum(axis =1)
```

```python
Tor_venues_grouped.sort_values('Total',ascending = False, inplace = True)
```

```python
Tor_venues_grouped_top20 = Tor_venues_grouped.iloc[0:20,0:-1]
```

```python
Tor_venues_grouped_top20[Tor_venues_grouped_top20.columns[2]].sum()
```

```python
a = []
for i in range(len(Tor_venues_grouped_top20.columns)):
    if Tor_venues_grouped_top20[Tor_venues_grouped_top20.columns[i]].sum()>20:
        a.append(Tor_venues_grouped_top20.columns[i])
    else:
        pass
```

```python
a
```

```python
Tor_venues_grouped_top20_Filtered = Tor_venues_grouped_top20[a]
```

```python
y = Tor_venues_grouped_top20_Filtered['Coffee Shop'].values
```

```python
Xfeature = Tor_venues_grouped_top20_Filtered.loc[:,Tor_venues_grouped_top20_Filtered.columns != 'Coffee Shop']
```

```python
from sklearn import preprocessing
X = preprocessing.StandardScaler().fit(Xfeature).transform(Xfeature.astype(float))
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
regr.fit(X,y)
yhat = regr.predict(X)
```

```python
regr.score(X,y)
```

```python
yhat
```

```python
yhat.round(2)
```

```python
difference= yhat.round(2) - y
```

```python
difference.round(2)
```

```python
dic = pd.DataFrame(Tor_venues_grouped_top20_Filtered.index)
```

```python
dic['Difference'] = difference.round(1)
```
